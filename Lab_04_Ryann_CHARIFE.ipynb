```python
import numpy as np
# Define 3 points in 2D-space:
X = np.array([[2, 1, 0], [4, 3, 0]])

# Calculate the covariance matrix:
R = np.matmul(X, X.T) / 3
print("Covariance Matrix:")
print(R)

# Alternative way to calculate the covariance matrix:
R_alternative = np.cov(X)
print("Alternative Covariance Matrix:")
print(R_alternative)

# Perform Singular Value Decomposition (SVD) to get new basis vectors:
U, D, V = np.linalg.svd(R)
u1 = U[:, 0]  # New basis vectors
u2 = U[:, 1]
print("New Basis Vectors:")
print(u1)
print(u2)

# Calculate the coordinates in the new orthonormal basis:
coordinates_point1 = np.dot(u1, X[:, 0]), np.dot(u2, X[:, 0])
coordinates_point2 = np.dot(u1, X[:, 1]), np.dot(u2, X[:, 1])
coordinates_point3 = np.dot(u1, X[:, 2]), np.dot(u2, X[:, 2])
print("Coordinates in New Basis:")
print("Point 1:", coordinates_point1)
print("Point 2:", coordinates_point2)
print("Point 3:", coordinates_point3)

# Reconstruct from the new basis:
X_reconstructed = np.dot(U.T, np.dot(U, X))
print("Reconstructed from New Basis:")
print(X_reconstructed)
```
